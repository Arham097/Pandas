### `Q-16:` Use the football dataset. Find out the total percentages that each team made on target. Display the result as a python dictionary where the keys are the team list and the values are the percentage values. Round off the percentage values up to 2 decimal places.

*Help:*
- First, find out how many total teams are participated in this worldcup. For that, you can use `unique()` method on the column "Team" or "Against".
- Loop through the teams list that you have found in previous section, and then filter the dataset according to that. After filtering the dataset, find out total attempts sum and total on target sum.
- After getting these values, find out the percentage by total on target divided by total attempts and multiply by 100. And store to a python dictionary where the key will be the team name and the values will be the percentages. 
- At the end,sort the dictionary by the values (not by the keys) and print the result.


````python
import pandas as pd

# Load the dataset
# df = pd.read_csv("football.csv")  # Replace with the correct filename

# Get a unique list of teams
teams = pd.unique(df[['Team', 'Against']].values.ravel())

# Dictionary to store team-wise on-target percentage
team_on_target_percentage = {}

# Loop through each team
for team in teams:
    team_data = df[df['Team'] == team]
    
    total_attempts = team_data['Total Attempts'].sum()
    total_on_target = team_data['On Target'].sum()
    
    # Calculate percentage (avoid division by zero)
    if total_attempts > 0:
        percentage = (total_on_target / total_attempts) * 100
    else:
        percentage = 0
    
    team_on_target_percentage[team] = round(percentage, 2)

# Sort dictionary by values in descending order
sorted_team_percentage = (sorted(team_on_target_percentage.items(), key=lambda x: x[1], reverse=True))

sorted_dict = {}
for x,y in sorted_team_percentage:
    sorted_dict.update([(x,y)])


# Print the sorted result
print(sorted_dict)
````

### **`Q-20:` Make a dataframe of each team in IPL with details like - Team Name, Matches Played, Win%, Home Win%, Away Win%.**
Show sorted dataframe on Win%

Replace old team name as new name before performing any tasks.
```
Delhi Daredevils ->Delhi Capitals
Kings XI Punjab -> Punjab Kings
Rising Pune Supergiants -> Rising Pune Supergiant
```

Note: Team1 represents Home team. Exclude not result matches.


````python
import pandas as pd

# Load the dataset
ipl = pd.read_csv('ipl-matches.csv')

# Replace old team names with new ones
team_name_replacements = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Rising Pune Supergiants': 'Rising Pune Supergiant'
}
ipl.replace(team_name_replacements, inplace=True)

# Exclude matches with no result
ipl = ipl[ipl['WinningTeam'].notna()]

# Get the list of unique teams
teams = pd.concat([ipl['Team1'], df['Team2']]).unique()

data = []

for team in teams:
    # Filter matches involving the current team
    team_matches = ipl[(ipl['Team1'] == team) | (ipl['Team2'] == team)]
    total_matches = len(team_matches)

    # Calculate wins
    total_wins = len(team_matches[team_matches['WinningTeam'] == team])
    win_percentage = (total_wins / total_matches) * 100 if total_matches else 0

    # Home wins
    home_matches = df[df['Team1'] == team]
    home_wins = len(home_matches[home_matches['WinningTeam'] == team])
    home_win_percentage = (home_wins / len(home_matches)) * 100 if len(home_matches) else 0

    # Away wins
    away_matches = df[df['Team2'] == team]
    away_wins = len(away_matches[away_matches['WinningTeam'] == team])
    away_win_percentage = (away_wins / len(away_matches)) * 100 if len(away_matches) else 0

    # Append team data
    data.append([team, total_matches, win_percentage, home_win_percentage, away_win_percentage])

# Create the final dataframe
team_stats = pd.DataFrame(data, columns=['Team Name', 'Matches Played', 'Win%', 'Home Win%', 'Away Win%'])

# Sort by Win% in descending order
team_stats = team_stats.sort_values(by='Win%', ascending=False).reset_index(drop=True)

team_stats
````

### **`Q-22:` IPL Point Table**

Make a function `point_table` which take `season` as parameter and show points table in non-ascendng order of points and in ascending order of team name.

For winning - 2 Ponits; 
For loosing - 0 Point
For not result both team gets 1 points.

Make dataframe which will have
`TeamName` `MatchesPlayed` `MatchesWon` `NoResult` `Points`
make `TeamName` as index.

```
season parametr should be one of these->
['2022', '2021', '2020/21', '2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011', '2009/10', '2009', '2007/08']
```



````python
def points_table(season):
  temp_season = ipl[ipl['Season']==season]
  teams = pd.concat([temp_season['Team1'],temp_season['Team2']]).unique()
  data = []
  for i in teams:
    played = len(temp_season[(temp_season['Team1'] == i) | (temp_season['Team2'] == i)])
    won = len(temp_season[(temp_season['WinningTeam'] == i)])
    no_result = temp_season[(temp_season['Team1'] == i) | (temp_season['Team2'] == i)]['WinningTeam'].isnull().sum()
    points = won*2
    data.append([i,played,won,no_result,points])
  temp_df = pd.DataFrame(data,columns=['TeamName','MatchesPlayed','MatchesWon', 'NoResult','Points']).sort_values(by='Points',ascending=False).reset_index(drop=True).set_index('TeamName')
  return temp_df
  
points_table('2022')
````

````python
def z_normalize(group,cols_to_perform):
  for col in cols_to_perform:
    group[f'{col}_z'] = ((group[col] - group[col].mean() )/ group[col].std() if group[col].std()!=0 else 0) 

    group['Min Passes'] = group['Passes'].min()
    group['Max Passes'] = group['Passes'].max()
    group['Min Yellow Cards'] = group['Yellow Cards'].min()
    group['Max Yellow Cards'] = group['Yellow Cards'].max()
    group['Average Yellow Cards'] = group['Yellow Cards'].mean()
    group['Max Attempted Line Breaks'] = group['Attempted Line Breaks'].max()
    group['Min Attempted Line Breaks'] = group['Attempted Line Breaks'].min()
    group['Std Attempted Line Breaks'] = group['Attempted Line Breaks'].std()
    group['Average Possession'] = group['Possession (%)'].mean()

    return group

teams = fifa_df.groupby('Team')
teams.apply(z_normalize,cols_to_perform=['Passes', 'Passes Completed','Attempted Line Breaks','Completed Line Breaks'])
````

### **`Q-56` Using both dataset, make a dataframe as described below**

Data Frame columns-> `['PlayerOfThematch', 'BattingFigure', 'BowlingFigure']`

* BattingFigure->`<runs>/<balls>`
* BowlingFigure->`<wicket>/<runs-conceded>`

DataFrame should have one record for each match.

Say 'V Kohli' got POM award then in dataset include his batting figure of that match. Say he scored 112runs in 76 balls. And he hasn't bowled so Bowling Figure will be NaN
```
PlayerOfThematch BattingFigure BowlingFigure
V Kohli          112/76         nan  

```


````python
# Merge both datasets on match ID (id)
full_matches = ipl_matches.merge(ipl_deliveries, left_on='id', right_on='match_id')

# Batting stats for player of the match
bat_s = full_matches[full_matches['batsman'] == full_matches['player_of_match']]
batting_stats = bat_s.groupby(['id', 'player_of_match']).agg(
    batsman_runs=('batsman_runs', 'sum'),
    balls=('ball', 'count')
).reset_index()
batting_stats['BattingFigure'] = batting_stats['batsman_runs'].astype(str) + '/' + batting_stats['balls'].astype(str)

# Bowling stats for player of the match
bol_s = full_matches[full_matches['bowler'] == full_matches['player_of_match']]
bowling_stats = bol_s.groupby(['id', 'player_of_match']).agg(
    runs=('total_runs', 'sum'),
    wickets=('player_dismissed', lambda x: x.notna().sum())
).reset_index()
bowling_stats['BowlingFigure'] = bowling_stats['wickets'].astype(str) + '/' + bowling_stats['runs'].astype(str)

# Merge on both id and player_of_match
final = pd.merge(
    batting_stats[['id', 'player_of_match', 'BattingFigure']],
    bowling_stats[['id', 'player_of_match', 'BowlingFigure']],
    on=['id', 'player_of_match'],
    how='outer'  # handles cases where POM either batted or bowled only
)

# Rename column for final output
final.rename(columns={'player_of_match': 'PlayerOfTheMatch'}, inplace=True)

# Optional: sort
final = final.sort_values(by='id')

# Preview
final
````
